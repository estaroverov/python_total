```python
from sympy import *
```

```python
from sympy.plotting import plot
```

```python
x = Symbol('x')
y = -5*x**2 + 10*x + 150
```

```python
plot(y, (x,-20,20))
```

```python
x1,x2 = solve(y,x) # Корни уравнения
```

```python
round(x1.evalf(),2)
```

```python
round(x2.evalf(),2)
```

```python
solve(diff(y)) # Вершина
```

```python
solve(y>0)# f>0
```

```python
solve(y<0)# f<0
```

```python
solve(diff(y)>0)# Функция возрастает
```

```python
solve(diff(y)<0)# Функция убывает
```

```python
plot(diff(y))
```
